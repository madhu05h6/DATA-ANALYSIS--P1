```python
import requests
```

```python
data= requests.get("https://visaguide.world/africa/")
data
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(data.text,"html.parser")
soup
```

```python
x=soup.find_all('table')
x[1]
```

```python
y=x[1].find_all('th')
z=[i.text.strip() for i in y]
z
```

```python
rowdata=x[1].find_all('td')
rowdata=[i.text.strip() for i in rowdata]
rowdata[:12]
```

```python
d = {header: [] for header in z}
for i in range(len(rowdata)):
  if i % 4==0:
    d['Country'].append(rowdata[i])
  if i % 4==1:
    d['Capital'].append(rowdata[i])
  if i % 4==2:
    d['Area km2'].append(rowdata[i])
  if i % 4==3:
    d['Population'].append(rowdata[i])
```

```python
import pandas as pd
df=pd.DataFrame(d)
df
```

```python
df["Area km2"]=df['Area km2'].str.replace(",",'').astype(int)
```

```python
df.sort_values(by='Area km2',ascending=False,inplace=True)
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(50,20))
x=df['Country']
y=df['Area km2']
plt.xticks(rotation=90,fontsize=30)
plt.yticks(fontsize=40)
plt.bar(x,y)
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
gdpr=requests.get("https://en.wikipedia.org/wiki/List_of_African_countries_by_GDP_(nominal)")
gdpSoup=BeautifulSoup(gdpr.text,"html.parser")
t=gdpSoup.find_all('table')
table=t[0]
h=table.find_all('th')
h=[i.text.strip() for i in h]
h
headings=h[:5]
headings
```

```python
rw=table.find_all('td')
rw=[i.text.strip() for i in rw]
rw[:12]
```

```python
rowdata=[]
for i in range(len(headings)):
  x=rw[i::len(headings)]
  rowdata.append(x)
len(rowdata)
```

```python
d={}
for i in range(len(rowdata)):
  d[headings[i]]=rowdata[i]
d
```

```python
import pandas as pd
dfgdp=pd.DataFrame(d)
dfgdp
```
